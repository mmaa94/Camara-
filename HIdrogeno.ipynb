```python
!head -n 5 1.dat
data1 = np.loadtxt('1.dat')
print type(data1)
print len(data1)
```

```python
def guardardatos(dato):

    #Arreglo que guarda los arreglos con la información de cada electrodo
    valores=np.empty([2,len(dato)])

    #Ciclo para guardar la información en cada arreglo
        
    for i in range(2):
            
        for j in range(len(dato)):

                (valores[i])[j]=(dato[j])[i]

    return valores

   
```

```python
one=guardardatos(data1)

for i in range(len(one[1])):
    
    print one[1][i]
```
